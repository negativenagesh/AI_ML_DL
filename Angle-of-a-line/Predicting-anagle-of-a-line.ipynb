```python
import pandas as p

traindata=p.read_csv("train.csv")
testdata=p.read_csv("test.csv")

traindata.head(25)
```

```python
x=traindata.drop(columns=['angle'])
y=traindata['angle']
```

```python
print(x.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(6, 6))

for i,ax in enumerate(axes.flat):
    ax.imshow(x.iloc[i].values.reshape(28, 28), cmap='gray')
    ax.axis('off')
plt.show()
```

# StandardScaler


```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(x)
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x=pca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train_scal=scaler.transform(x_train)
x_test_scal=scaler.transform(x_test)
```

```python
x_train_scal=p.DataFrame(x_train_scal,columns=x_train.columns)
x_test_scal=p.DataFrame(x_test_scal,columns=x_test.columns)
```

```python
x_train_scal.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(6, 6))

for i,ax in enumerate(axes.flat):
    ax.imshow(x_train_scal.iloc[i].values.reshape(28, 28), cmap='gray')
    ax.axis('off')
plt.show()
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)

from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
