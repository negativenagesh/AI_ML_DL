# 1. Loadind data 

```python
import pandas as pd
```

```python
data=pd.read_csv('sonar.csv',header=None)
data.head(10)
```

```python
data_features=data.drop(columns=[60])
data_features.head()
```

```python
target=data[60]
target.head()
```

# 2. Histogram of features(60)

```python
import seaborn as sea
import matplotlib.pyplot as plt
for column in range(60):
    sea.histplot(data=data[column])
    plt.show()
```

# 3. Applying Box-Cox 

```python
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler(feature_range=(1,2))
scaled_data_features=pd.DataFrame(scale.fit_transform(data_features))

# scaled_data_features=scale.fit(data_features)
# scaled_data_features=scale.transform(scaled_data_features)
# transformed_data_features=scale.transform([[1,2]])
# data_features=pd.DataFrame(scale.transform(data_features))
```

```python
from sklearn.preprocessing import power_transform

power_tran=pd.DataFrame(power_transform(scaled_data_features,method='box-cox'))
```

# 4. Histogram of transformed features

```python
for column in range(60):
    sea.histplot(data=power_tran[column])
    plt.show()
```

# 5. PCA, plotting pc1 and pc2 using seaborn scatterplot

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
power_trans1=pca.fit(power_tran)
```

```python
power_trans1.components_
```

```python

pcaeddata=pd.DataFrame(data=power_trans1.components_.T,columns=['pc1','pc2'])
```

```python
sea.scatterplot(data=pcaeddata)
plt.show()
```
